3. Определить метки времени, когда есть вокал.

```python
dd0, sr =torchaudio.load('/content/audio_only/output/audio_pugacheva/vocals_mono.wav')
```

```python
# ищем уровень тишины
sul=[]
b_size = 11025
for i in range(0,dd0[0].shape[0],b_size):
    sul.append(torch.abs(dd0[0,i:(i+b_size)]).sum())


treshold = torch.quantile(torch.tensor(sul),0.25).item()
treshold
```

```python
labels = []
is_speach = False
start = 0
for i in range(0,dd0[0].shape[0],b_size//3):
    if torch.abs(dd0[0,i:(i+b_size)]).sum()>treshold:
        if is_speach == False:
            start = i
            is_speach = True
    else:
        if is_speach:
            if i-start > b_size*3: #берем отрезки больше 1 секунды
                labels.append((start,i))
                is_speach = False
            else:
                start = i
```

```python
print('количество сегментов: ', len(labels))
```

```python
print('Длительности сегментов в секундах: ',[(x[1]-x[0])//44100 for x in labels])
```

4. Распознать текст песни с помощью открытых библиотек распознавания речи и определить метрики качества распознавания текста из аудио.¶
делим файл на короткие чанки поскольку speech_recognition имеет ограничения по размеру данных

```python
start = 0
ss = torch.tensor([])
k = 0
for x in labels:
    ss = torch.hstack([ss, dd0[0][x[0]:x[1]]])
    if ss.shape[0]//44100 > 70:
        fname = 'audio_pugacheva_vocals_mono_' + str(k) + '.wav'
        print(fname,'// длительность:',ss.shape[0]//44100)
        torchaudio.save(fname,ss.unsqueeze(0),sample_rate=44100)
        start = x[1] + 1
        k += 1
        soun_ = AudioSegment.from_wav(fname)
        soun_.export(fname, format='wav')
        ss = torch.tensor([])
        
if ss.shape[0] !=0 :        
    fname = 'audio_pugacheva_vocals_mono_' + str(k) + '.wav'
    print(fname,'// длительность:',ss.shape[0]//44100)
    torchaudio.save(fname,ss.unsqueeze(0),sample_rate=44100)
    soun_ = AudioSegment.from_wav(fname)
    soun_.export(fname, format='wav')
    k +=1
```
